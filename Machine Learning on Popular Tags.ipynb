```python
# CSV file loaded 
file_to_load = './PopularTags_DataSet.csv'
```

```python
# Import our dependencies 
import pandas as pd 
import numpy as np 
```

```python
popular_tags = pd.read_csv(file_to_load)

popular_tags.head()
```

```python
popular_tags = popular_tags.rename(columns = {"Tag_Name": "game_name"})
```

```python
popular_tags = popular_tags.drop_duplicates()
```

```python
popular_tags = popular_tags.drop_duplicates(subset=['game_name'])
```

```python
popular_tags = popular_tags.dropna()
```

```python
popular_tags.to_csv(r'./PopularTags_DataSet.csv', index = False)
```
